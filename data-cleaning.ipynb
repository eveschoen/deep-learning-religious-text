# Generate Tokens and Vocab tables

```python
import nltk

nltk_resources = [
    'punkt', 
    'averaged_perceptron_tagger', 
    'stopwords', 
    'tagsets',
    'wordnet'
]

for rsc in nltk_resources:
    nltk.download(rsc)
    
from nltk.corpus import stopwords
```

## Bible

```python
bible.head()
```

```python
# remove punctuation from text - "Clean Text" column
bible["Clean Text"] = bible['Text'].str.replace('[^\w\s]','')
```

```python
bible.head()
```

```python
# generate pos, tokens, terms
bible_tokens = bible['Clean Text'].apply(lambda x: pd.Series(nltk.pos_tag(nltk.word_tokenize(x)))).stack().to_frame('pos_tuple')

bible_tokens['pos'] = bible_tokens.pos_tuple.apply(lambda x: x[1])
bible_tokens['token'] = bible_tokens.pos_tuple.apply(lambda x: x[0])
bible_tokens['term'] = bible_tokens.token.str.lower()

# lemmatize
lemmatizer = nltk.stem.WordNetLemmatizer()

lemmatized = []
for word in bible_tokens['term']:
    lemmatized.append(lemmatizer.lemmatize(word))
    
bible_tokens['lemma'] = lemmatized   
```

```python
bible_tokens.head(10)
```

```python
#bible_tokens.reset_index().to_csv(f'{file_path}/bible_tokens.csv', index=False)
```

```python
bible_vocab = bible_tokens.groupby(level=0)['lemma'].value_counts().to_frame('n')
bible_vocab = bible_vocab.reset_index()
bible_vocab = bible_vocab.set_index('lemma')

# add stopwords
new_stopwords = ["unto", "said", "thy", "thou", "thee", "shall", "upon", "ye", "saying", "saith", "hath", "forth", "also", "began", "put", "wa"]
stpwrd = nltk.corpus.stopwords.words('english')
stpwrd.extend(new_stopwords)

#sw = pd.DataFrame(nltk.corpus.stopwords.words('english'), columns=['term'])
sw = pd.DataFrame(stpwrd, columns=['lemma'])
sw = sw. reset_index().set_index('lemma')
sw.columns = ['dummy']
sw.dummy = 1

bible_vocab['stop'] = bible_vocab.index.map(sw.dummy)
bible_vocab['stop'] = bible_vocab['stop'].fillna(0).astype(int)

# drop stopwords
bible_vocab = bible_vocab[bible_vocab['stop'] != 1].drop('stop', axis=1).reset_index().set_index(['Book Name', 'lemma'])
```

## Buddha

```python
# remove punctuation from text - "Clean Text" column
buddha["Clean Text"] = buddha['translation-en-sujato'].str.replace('[^\w\s]','')
```

```python
buddha.head()
```

```python
# generate pos, tokens, terms
buddha_tokens = buddha['Clean Text'].apply(lambda x: pd.Series(nltk.pos_tag(nltk.word_tokenize(str(x))))).stack().to_frame('pos_tuple')

buddha_tokens['pos'] = buddha_tokens.pos_tuple.apply(lambda x: x[1])
buddha_tokens['token'] = buddha_tokens.pos_tuple.apply(lambda x: x[0])
buddha_tokens['term'] = buddha_tokens.token.str.lower()

# lemmatize
lemmatizer = nltk.stem.WordNetLemmatizer()

lemmatized = []
for word in buddha_tokens['term']:
    lemmatized.append(lemmatizer.lemmatize(word))
    
buddha_tokens['lemma'] = lemmatized   
```

```python
buddha_tokens.head()
```

```python
#buddha_tokens.reset_index(level=4, drop=True).reset_index().to_csv(f'{file_path}/buddha_tokens.csv', index=False)
```

```python
buddha_vocab = buddha_tokens.groupby(level=0)['lemma'].value_counts().to_frame('n')
buddha_vocab = buddha_vocab.reset_index()
buddha_vocab = buddha_vocab[buddha_vocab['lemma'] != 'nan']
buddha_vocab = buddha_vocab.set_index('lemma')

# add stopwords
new_stopwords = ["unto", "said", "thy", "thou", "thee", "shall", "upon", "ye", "saying", "saith", "hath", "forth", "also", "began", "put", "wa"]
stpwrd = nltk.corpus.stopwords.words('english')
stpwrd.extend(new_stopwords)

#sw = pd.DataFrame(nltk.corpus.stopwords.words('english'), columns=['term'])
sw = pd.DataFrame(stpwrd, columns=['lemma'])
sw = sw.reset_index().set_index('lemma')
sw.columns = ['dummy']
sw.dummy = 1

buddha_vocab['stop'] = buddha_vocab.index.map(sw.dummy)
buddha_vocab['stop'] = buddha_vocab['stop'].fillna(0).astype(int)

# drop stopwords
buddha_vocab = buddha_vocab[buddha_vocab['stop'] != 1].drop('stop', axis=1).reset_index().set_index(['level_1', 'lemma'])
```
